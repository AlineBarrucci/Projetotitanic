```python
#The objective of the project is to create a model that predicts how many passengers survived the Titanic wreck.
#import of the libraries to be used
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
#Reading Test Base and Training Base.
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
#5 first lines of the training set
train.head()
```

```python
train.drop(['Name','Ticket','Cabin'],axis=1,inplace=True)
test.drop(['Name','Ticket','Cabin'], axis=1,inplace=True)
```

```python
#Transformation of data into numerical values.
new_data_train = pd.get_dummies(train)
new_data_test = pd.get_dummies(test)
```

```python
#New test base
new_data_test.head()
```

```python
#New training base
new_data_train.head()
```

```python
#identifying the number of null data in the training base
new_data_train.isnull().sum().sort_values(ascending=False).head()
```

```python
#inserting the mean age in place of the null values.
new_data_train['Age'].fillna(new_data_train['Age'].mean(), inplace=True)
new_data_test['Age'].fillna(new_data_test['Age'].mean(), inplace= True)
```

```python
#inserting the mean age in place of the null values.
new_data_test.isnull().sum().sort_values(ascending=False).head(10)
new_data_test['Fare'].fillna(new_data_test['Fare'].mean(),inplace=True)
```

```python
#Separating the learning data (x) e the data that we want to predict (y)
x = new_data_train.drop('Survived', axis=1)
y = new_data_train['Survived']
print(new_data_train['Survived'])
```

```python
#Model"Decision Tree"
tree = DecisionTreeClassifier(max_depth=3,random_state=0)
tree.fit(x,y)

#Accuracy level model Decision Tree
pre = tree.score(x,y)
print(pre)
```
